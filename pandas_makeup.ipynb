Merge/Concatenate `df_mexico_2` and `df_mexico_3` and save it to a new dataframe `df_new`

```python
import pandas as pd
```

```python
df_mexico_2 = pd.read_csv("data/mexico-city-real-estate-2.csv")
df_mexico_3 = pd.read_csv("data/mexico-city-real-estate-3.csv")
print("df_mexico_2 shape:", df_mexico_2.shape)
print("df_mexico_3 shape:", df_mexico_3.shape)
```

```python
df_new = pd.concat([df_mexico_2, df_mexico_3])
print("df_new shape:", df_new.shape)
df_new.head()
```
